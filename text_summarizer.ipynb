```python
# importing libraries
from bs4 import BeautifulSoup
import urllib.request
import nltk
nltk.download('stopwords')
```

```python
# getting data

#data = input("Enter the topic's summary you want from Wikipedia: ").split(" ")
#data = "_".join(data)
data = "indian_economy"
source = urllib.request.urlopen("https://en.wikipedia.org/wiki/"+data).read()
#print(source)
soup = BeautifulSoup(source,"lxml")
```

```python
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
text
```

```python
# preprocessing

import re

text = re.sub(r"\[[0-9]*\]"," ", text)
text = re.sub(r"\s+"," ",text)

sentences = nltk.sent_tokenize(text)

clean_text = text.lower()
clean_text = re.sub(r"\W"," ",clean_text)
clean_text = re.sub(r"\d"," ",clean_text)
clean_text = re.sub(r"\s+"," ",clean_text)

stop_words = nltk.corpus.stopwords.words('english')
```

```python
# histogram

word2count = {}
for word in nltk.word_tokenize(clean_text):
    if word not in stop_words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1
            
word2count
            
```

```python
# weighted histogram   [b/w 0 to 1]

for key in word2count.keys():
    word2count[key] = word2count[key]/max(word2count.values())
word2count
```

```python
# sentence score
sent2core = {}

for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word2count.keys():
            # for less than 15 words in a sentence
            if len(sentence.split(" "))<15:
                if sentence not in sent2core.keys():
                    sent2core[sentence] = word2count[word]
                else:
                    sent2core[sentence] += word2count[word]
                
sent2core
        
```

```python
import heapq

summary_sentences = heapq.nlargest(5,sent2core,key = sent2core.get)

for sentence in summary_sentences:
    print(sentence)
```
